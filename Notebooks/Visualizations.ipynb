# Mexico Plots of Infrastructure vs Total # of Cases

```python
mex.head()
```

```python
#creating a list of columns that we want to show the relation with number of viral cases over time.
columns = ['ICT goods exports (% of total goods exports)',
       'Individuals using the Internet (% of population)',
       'Fixed telephone subscriptions (per 100 people)',
       'Air transport, freight (million ton-km)',
       'Industrial design applications, nonresident, by count',
       'Public private partnerships investment in water and sanitation (current US$)',
       'Public private partnerships investment in transport (current US$)',
       'Electricity production from natural gas sources (% of total)',
       'ICT service exports (% of service exports, BoP)',
       'Healthcare Expenditure Percent GDP',
       'Pharmaceutical Spending US $ / Capita', 'CT Scan Device Counts',
       'PET Scan Device Counts', 'Doctors per 1000 People',
       'Nurses per 1000 People', 'Hospital Count']
```

```python
#for every column in column
for x in columns:
    
    #create a scatter plot of that column and total cases
    fig, ax = plt.subplots(figsize = (10,6))
    ax.scatter(mex[x], mex['Total Cases'])
    
    #setting x label to be the name of the column
    ax.set_xlabel(str(x), fontsize = 15)
    
    #our Y is our total cases of Pertussis and Mumps
    ax.set_ylabel('# of Cases of Pertussis & Mumps', fontsize = 15)
    
    #setting title based on iterated column
    plt.title(str(x) + ' vs # of Cases', fontsize = 15)
    
    plt.tick_params(labelsize=15)
    plt.show()
```

**Notes**: Some notable relationships:

**Social/Transportation**: There is a strong opsitive relationship between infrastructure that contributes to socializing and number of cases. For example, this is exhibited with telephone subscriptions, percentage of individuals using the internet, air transport, and transport investment. More social cities and countries exhibiting higher rates of transmission makes sense and an interesting area to look at.

**Sanitation/Electricity**: One would think as sanitation and electricity funding goes up then spread of disease and access to information or the modernizing of a society would help limit the spread of disease but the opposite is observed. However as shown below and from examing correlations we know that there is little correlation with sanitation and # of cases suprisingly.

**Healthcare Infrastructure**: Again, the opposite of our expectations is shown. As healthcare infrastructure metrics such as spending, # of doctors, # of hospitals, # medical technology increases so do the number of diseases. Leads me to believe this is an error in causality, a which came first situation. Perhaps the increase of these things is a function of disease, i.e., the more cases the more a country allocates resoureces towards healthcare, instead of adjusting healthcare funding before outbreaks, government is more reactionary than preparatory.




# Mexico Infrastructure Heatmap
Reference: https://towardsdatascience.com/better-heatmaps-and-correlation-matrix-plots-in-python-41445d0f2bec

```python
mex_heat = mex.drop(axis = 0, columns = ['Year'], inplace = True)
```

```python
sns.set(font_scale=5)
corr = mex.corr()
fig, ax = plt.subplots(figsize=(50,50))
ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=70,
    horizontalalignment='right'
);
```
